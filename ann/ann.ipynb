```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.10 * np.random.randn(n_inputs, n_neurons)
        display(self.weights.shape)
        self.biases = np.zeros((1,n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
        
class Activation_ReLU:
    def forward(self, inputs):
        self.output = np.maximum(0, inputs)

inputs = inputs.reshape(99,1) #wow ok so apparently this operation does not act on the object...not a persistent change, one must reassign to an l-value
inputs.shape
#display(inputs.shape, inputs)
layer1 = Layer_Dense(1,5)
layer2 = Layer_Dense(5,1)
activation1 = Activation_ReLU()
activation2 = Activation_ReLU()

layer1.forward(inputs)
print(layer1.output)
activation1.forward(layer1.output)

# layer2.forward(layer1.output)
# print(layer2.output)
```
